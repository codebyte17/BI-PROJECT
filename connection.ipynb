## Inserting the ANONDIM data

```python
# Read data from the CSV file
data = pd.read_csv("ANONIDDIM.csv")
```

```python
# Loop through each row in the DataFrame and insert into the ANONDIM table
for index, row in data.iterrows():
    # Convert the types explicitly to native Python types
    id_value = int(row['ID'])  # Convert numpy.int64 to Python int
    anonid_value = str(row['ANONID'])  # Ensure it's a string

    # Prepare the insert statement
    insert_query = """
    INSERT INTO anondim (ID, ANONID) 
    VALUES (%s, %s)
    """
    
    # Prepare the data tuple
    data_tuple = (id_value, anonid_value)
    
    # Execute the insert statement
    cur.execute(insert_query, data_tuple)

# Commit the transaction
conn.commit()
```

## Load TIMEDIM data

```python
timedim = pd.read_csv("TIMEDIM.csv")
```

```python
timedim.head(10)
```

```python
# Loop through each row in the DataFrame and insert into the ANONDIM table
for index, row in timedim.iterrows():
    # Convert the types explicitly to native Python types
    ID = int(row["ID"])
    year = int(row['year'])
    dict ={
            "march":3,
            "april":4,
            "may":5
          }
    month = int(dict[str(row['month']).strip()])
    weekday = str(row['weekday'])
    day_of_month = str(row['day of the month'])
    
    # Prepare the insert statement
    insert_query = """
    INSERT INTO TIMEDIM (ID, year,month,weekday,day_of_month) 
    VALUES (%s, %s,%s,%s,%s)
    """
    
    # Prepare the data tuple
    data_tuple = (ID, year,month,weekday,day_of_month)

    cur.execute(insert_query, data_tuple)
   

# Commit the transaction
conn.commit()
```

## FACTS loading to DB

```python
facts_df = pd.read_csv("FACTS.csv")
```

```python
facts_df = facts_df[900000:4000000]
```

```python
# Loop through each row in the DataFrame and insert into the ANONDIM table
for index, row in facts_df.iterrows():
    # Convert the types explicitly to native Python types
    
    anonid = int(row['ANONID'])
    timeid = int(row['TIMEID'])
    queryid = int(row['QUERYID']) 
    urlid = int(row['URLID'])
    click = int(row['CLICK'])
    irank = float(row['IRANK'])
    
    # Prepare the insert statement
    insert_query = """
    INSERT INTO FACTS (ID,ANONID,QUERYID,TIMEID,URLID,IRANK,CLICK) 
    VALUES (%s, %s,%s,%s,%s,%s,%s)
    """
    
    # Prepare the data tuple
    data_tuple = (index+1,anonid,queryid,timeid,urlid,irank,click)
    
    # Execute the insert statement
    cur.execute(insert_query, data_tuple)

# Commit the transaction
conn.commit()
```
